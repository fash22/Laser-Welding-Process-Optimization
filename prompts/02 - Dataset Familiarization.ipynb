```python
from IPython.display import Markdown

import os
import openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key  = os.environ['OPENAI_API_KEY']

def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0, max_tokens=1000):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response.choices[0].message["content"]
```

```python
instructions = """
You will be helping me in writing an article about the "Screening datasets for laser welded steel-copper lap joints" dataset.
This dataset can be downloaded from https://data.mendeley.com/datasets/2s5m3crbkd/2.
Do not assume that 
The article should be short, concise.
The tone should be professional, academic, technical, but compelling.

Return your results in markdown
"""

generate_article = """
Create an article that contains the following.

- Where to download V1, V1.1, V2, V2.1 dataset.
- What are their file types, make sure to read https://data.mendeley.com/datasets/2s5m3crbkd/2 before answering this one. DO NOT ASSUME.
- How to load the dataset in pandas knowing their file types.
"""

messages = [
    {
        "role": "system",
        "content": instructions
    },
    {
        "role": "user",
        "content": generate_article
    },
]

results = get_completion_from_messages(messages=messages)
Markdown(results)
```
